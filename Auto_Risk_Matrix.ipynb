```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=[10,10])
plt.subplots_adjust(wspace=0, hspace=0)
plt.xticks([])
plt.yticks([])
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.xlabel('Impact')
plt.ylabel('Likelihood')
plt.title('Risk Matrix')

axes = [fig.add_subplot(5, 5, z + 1) for z in range(5 * 5)]
for ax in axes:
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlim(0,5)
    ax.set_ylim(0,5)

red = [1,2,3,4,7,8, 9, 14]
yellow = [0, 5, 6, 11, 17, 23, 12, 13, 18, 19, 24]
green = [10, 15, 16, 20, 21, 22]

for _ in green:
    axes[_].set_facecolor('green')
for _ in yellow:
    axes[_].set_facecolor('yellow')
for _ in red:
    axes[_].set_facecolor('red')

items = [
    {'id': '03', 'name': 'Power outage', 'L': 1, 'I': 4},
    {'id': '07', 'name': 'Lift broke down', 'L': 2, 'I': 3},
    {'id': '01', 'name': 'AC failure', 'L': 2, 'I': 3},
    {'id': '04', 'name': 'Sensor error', 'L': 2, 'I': 3},
    {'id': '02', 'name': 'Water leak', 'L': 3, 'I': 1},
    {'id': '06', 'name': 'COVID-19', 'L': 4, 'I': 2},
    {'id': '05', 'name': 'Data corruption', 'L': 4, 'I': 3},
    {'id': '05', 'name': 'Fire', 'L': 4, 'I': 3},
]
risk_cnt = np.zeros((5, 5), dtype=int)
risk_plotted = np.zeros((5, 5), dtype=int)
for item in items:
    risk_cnt[item['L'] - 1][item['I'] - 1] += 1

for item in items:
    idx = item['L'] - 1
    jdx = item['I'] - 1
    box_idx = 5 * (4 - idx) + jdx
    axes[box_idx].text(0.5, 4 - risk_plotted[idx][jdx] / risk_cnt[idx][jdx] * 5, item['name'], bbox={'facecolor': 'white', 'alpha': 0.85, 'pad': 3})
    risk_plotted[idx][jdx] += 1


plt.show()
```
